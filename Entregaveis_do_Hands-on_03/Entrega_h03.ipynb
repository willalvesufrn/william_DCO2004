## Parte 1: Plote do espectro dos arquivos Violino.wav e Flauta.wav

```matlab
close all;clc;clear all;

% Abrindo os arquivos de √°udio (vt: amplitude das amostras, fs: frequ√™ncia de amostragem
[vtFlauta, fsFlauta] = audioread('/home/labsim/william_DCO2004/Entregaveis_do_Hands-on_03/Flauta.wav');
[vtViolino, fsViolino] = audioread('/home/labsim/william_DCO2004/Entregaveis_do_Hands-on_03/Violino.wav');

% Gr√°fico 1: Espectro de Flauta.wav
subplot(2,1,1)
fFlauta = linspace(-fsFlauta/2, fsFlauta/2, size(vtFlauta,1));
FFlauta = fftshift(abs(fft(vtFlauta)))/length(fFlauta);
plot(fFlauta, FFlauta);
axis([0 1e4 0 0.06])
title('Espectro de Flauta.wav');
xlabel('Frequ√™ncia(Hz)');
ylabel('|Amplitude|');

% Gr√°fico 2: Espectro de Violino.wav
subplot(2,1,2)
fViolino = linspace(-fsViolino/2, fsViolino/2, size(vtViolino,1));
FViolino = fftshift(abs(fft(vtViolino)))/length(fViolino);
plot(fViolino, FViolino);
axis([0 1e4 0 0.04])
title('Espectro de Violino.wav');
xlabel('Frequ√™ncia(Hz)');
ylabel('|Amplitude|');

```

## Parte 3: Mostra que voc√™ consegue obter resultados similares no Matlab e no Python.

```matlab
import numpy as np
from numpy.fft import *
import scipy.io.wavfile as wv 
import matplotlib.pyplot as plt;
%matplotlib notebook

# Abrindo arquivos
fsFlauta,vtFlauta = wv.read('/home/labsim/william_DCO2004/Entregaveis_do_Hands-on_03/Flauta.wav')
fsViolino,vtViolino = wv.read('/home/labsim/william_DCO2004/Entregaveis_do_Hands-on_03/Violino.wav')

# Figura 1: Espectro de Flauta.wav
plt.figure(1)
fFlauta = np.linspace(-fsFlauta/2, fsFlauta/2, np.size(vtFlauta))
FFlauta = fftshift(abs(fft(vtFlauta)))/np.size(fFlauta)
plt.plot(fFlauta, FFlauta)
plt.xlim([0, 1e4])
plt.title('Espectro de Flauta.wav')
plt.xlabel('Frequ√™ncia(Hz)')
plt.ylabel('|Amplitude|')

# Figura 2: Espectro de Violino.wav
plt.figure(2)
fViolino = np.linspace(-fsViolino/2, fsViolino/2, np.size(vtViolino))
FViolino = fftshift(abs(fft(vtViolino)))/np.size(fViolino)
plt.plot(fViolino, FViolino)
plt.xlim([0,1e4])
plt.title('Espectro de Violino.wav')
plt.xlabel('Frequ√™ncia(Hz)')
plt.ylabel('|Amplitude|')


```
